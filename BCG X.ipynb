```python
#Import Necessary Libraries
import pandas as pd
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import compiled data
df = pd.read_csv('10-K filings.csv')

df
```

```python
#Create new cloumns depicting percent change via groupby and pct_change functions
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100

df
```

<b>Microsoft:</b> Looking at the year-over-year changes for Microsoft, it's evident that revenue increased over both years. From 2022 to 2023 revenue increased by about 6.88%, and from 2023 to 2024 revenue increased by about 15.67%. However, the net income of Microsoft doesn't follow the same pattern. From 2022 to 2023 net income actually decreased by about 0.52%, but from 2023 to 2024 net income ended up increasing by a staggering 21.80%. Overall, revenua and growth both grew for Microsoft despite the slight decrease in net income over the first year.

<b>Tesla:</b> Looking at the year-over-year changes for Tesla, revenue also increased over both years. From 2022 to 2023 revenue increased by about 18.80%, and from 2023 to 2024 revenue increased by about 1.23%. However, the net income of Microsoft doesn't follow the same pattern. From 2022 to 2023 net income also increased by around by about 19.0%, but from 2023 to 2024 net income ended up decreasing by an enormous -52.23% despite revenue increasing that year. This may indicate an increased amount of spending during this period which lead to a sharp decrease in net income.

<b>Apple:</b> Looking at the year-over-year changes for Apple, from 2022 to 2023 revenue decreased by about 2.80%, and from 2023 to 2024 revenue increased by about 2.02%. However, the net income for Apple from 2022 to 2023 actually decreased by about -2.81%, but from 2023 to 2024 net income decreased by a similar margin of 3.36%. So, depite seeing smoe changes in revenue Apple didn't see income growth across either of these years

```python
#Create new columns depicting more trends via groupby and pct_change functions
df['Assest Growth (%)'] = df.groupby(['Company'])['Total Assets'].pct_change() * 100
df['Liability Growth (%)'] = df.groupby(['Company'])['Total Liabilities'].pct_change() * 100

df
```

<b>Microsoft:</b> From 2022 to 2023 Microsoft assets grew by 13.0% and from 2023 to 2024 their assets grew by 24.32%. Additionally, from 2022 to 2023 Microsoft liabilities grew by 3.76% and from 2023 to 2024 their liabilities grew by 18.52%. Overall, this supports the idea that the company grew over this time frame and notably gained more assets than it did liabilities.

<b>Tesla:</b> From 2022 to 2023 Tesla assets grew by 29.49% and from 2023 to 2024 their assets grew by 14.49%. Additionally, from 2022 to 2023 Tesla liabilities grew by 18.03% and from 2023 to 2024 their liabilities grew by 12.51%. Overall, this supports the idea that the company grew over this time frame and notably gained more assets than it did liabilities.

<b>Apple:</b> From 2022 to 2023 Apple assets shrank by 0.04% and from 2023 to 2024 their assets grew by 3.52%. Additionally, from 2022 to 2023 Microsoft liabilities shrank by 3.86% and from 2023 to 2024 their liabilities grew by 6.06%. Once again, this supports the idea that the company grew over this time frame and notably gained more assets than it did liabilities.

```python
#Calculate the yearly mean revenue across all companies each year
mean_revenue_2022 = np.round(df.query('Year == 2022')['Total Revenue'].mean(), decimals = 2)
mean_revenue_2023 = np.round(df.query('Year == 2023')['Total Revenue'].mean(), decimals = 2)
mean_revenue_2024 = np.round(df.query('Year == 2024')['Total Revenue'].mean(), decimals = 2)

display(mean_revenue_2022)
display(mean_revenue_2023)
display(mean_revenue_2024)

yearly_revenue = [mean_revenue_2022, mean_revenue_2023, mean_revenue_2024]
```

The above numbers represent the average revenue for all companies for the years of 2022, 2023, and 2024. As evident by the change in these numbers, the average revenue for companies increased each year.

```python
#Create a visulization depicting average revenue changes across years
sns.set_theme()

x = np.arange(3)
width = 0.5
values = yearly_revenue
colors = sns.color_palette("Greens", len(values))
categories = ['2022', '2023', '2024']

plt.figure(figsize = (6,5))

plt.bar(x, values, width, color = colors, edgecolor = 'black')

plt.xlabel('Year')
plt.ylabel('Average Total Revenue ($)')
plt.title('Average Total Revenue by Year', fontsize = 15)
plt.ylim(200000,250000)
plt.xticks(x, categories, size = 11)
plt.tight_layout()

plt.show()
```

The above visual further depicts how the average total revenue for all companies increased with each year. This is made evident by the increasingly taller bars present in the bara graph.

```python
#Calculate the yearly mean net income across all companies each year
mean_net_income_2022 = np.round(df.query('Year == 2022')['Net Income'].mean(), decimals = 2)
mean_net_income_2023 = np.round(df.query('Year == 2023')['Net Income'].mean(), decimals = 2)
mean_net_income_2024 = np.round(df.query('Year == 2024')['Net Income'].mean(), decimals = 2)

display(mean_net_income_2022)
display(mean_net_income_2023)
display(mean_net_income_2024)

yearly_net_income = [mean_net_income_2022, mean_net_income_2023, mean_net_income_2024,]
```

The above numbers represent the average net income for all companies for the years of 2022, 2023, and 2024. As evident by the change in these numbers, the companies saw and slight decrease in their net income from 2022 to 2023, but saw a decent increase in their net income from 2023 to 2024.

```python
#Create a visulization depicting average net income changes across years
sns.set_theme()

x = np.arange(3)
width = 0.5
values = yearly_net_income
colors = sns.color_palette("Blues", len(values))
categories = ['2022', '2023', '2024']

plt.figure(figsize = (6,5))

plt.bar(x, values, width, color = colors, edgecolor = 'black')

plt.xlabel('Year')
plt.ylabel('Average Net Income ($)')
plt.title('Average Net Income by Year', fontsize = 15)
plt.ylim(60000,65000)
plt.xticks(x, categories, size = 11)
plt.tight_layout()

plt.show()
```

The above visual further depicts how the average net income for all companies change with each year. This is made evident by the decrease in bar height over the first year, and the subsequent increase in bar height over the next year.
